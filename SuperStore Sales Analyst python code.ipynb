##1. Tổng quan doanh số và lợi nhuận của công ty theo từng năm

```python
#Thêm cột Year của OrderDate vào df
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Year'] = df['Order Date'].dt.year
df.head()
```

```python
df['Year'].unique()
```

```python
#Sắp xếp dữ liệu theo thời gian
df = df.sort_values(by='Order Date')
df.head()
```

```python
# ép kiểu cột Year
df['Year'] = df['Year'].astype(str)
df.head()
```

```python
#Gộp nhóm doanh thu, lợi nhuận, số lượng sản phẩm theo năm
df_year = df.groupby('Year')[['Sales','Profit','Quantity']].sum()
df_year
```

Sales (Doanh thu): Tổng doanh thu hàng năm có xu hướng tăng dần qua các năm:

- Năm 2014: $484,247.50

- Năm 2015: $470,532.51 (giảm nhẹ so với năm 2014)

- Năm 2016: $609,205.60 (tăng mạnh so với năm trước)

- Năm 2017: $733,215.26 (tiếp tục tăng đáng kể)

Profit (Lợi nhuận): Lợi nhuận cũng có xu hướng tăng qua các năm, phản ánh hiệu suất bán hàng tốt hơn hoặc cải thiện trong chi phí và giá bán.

Quantity (Số lượng): Số lượng bán hàng cũng tăng đều qua từng năm, cho thấy mức tăng trưởng về mặt số lượng sản phẩm được bán ra.


```python
#Vẽ biểu đồ biểu diễn doanh số và lợi nhuận của cửa hàng theo năm
plt.figure(figsize=(8,4))
sns.set_style('whitegrid')
sns.lineplot(data= df_year,x= 'Year', y= 'Sales', label='Sales', color='blue', marker = 'o')
sns.lineplot(data= df_year,x= 'Year', y= 'Profit', label='Profit', color='green', marker = 'o')
# Hiển thị biểu đồ
plt.title('Sales and Profit Over Years')
plt.xlabel('Year')
plt.ylabel('Sales Amount')
plt.legend()
plt.show()
```

Từ biểu đồ, ta thấy doanh số và lợi nhuận của cửa hàng SuperStore nhìn chung tăng dần qua các năm.

Từ năm 2015-2017, Doanh số có sự tăng trưởng mạnh từ 470532.5090 USD lên 733215.2552 USD (tăng hơn 55%), lợi nhuận cũng tăng từ 61618.6037 USD lên 93493.2696 USD.

##Phân tích ảnh hưởng của Discount đến Sales và Profit

```python
# Vẽ biểu đồ phân tán
plt.figure(figsize=(12, 6))

# Vẽ mối quan hệ giữa discount và sales
plt.subplot(1, 2, 1)
plt.scatter(df['Discount'], df['Sales'], color='blue')
plt.title('Discount vs Sales')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.grid()

# Vẽ biểu đồ mối quan hệ giữa discount và profit
plt.subplot(1, 2, 2)
plt.scatter(df['Discount'], df['Profit'], color='green')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid()

plt.tight_layout()
plt.show()
```

Discount vs Sales:

* Dữ liệu cho thấy khi mức discount tăng, sales (doanh thu) không có xu hướng tăng rõ rệt. Thay vào đó, sales cao có xu hướng xuất hiện ở mức discount thấp (khoảng từ 0 đến 0.2). Điều này có thể cho thấy việc tăng giảm giá không phải lúc nào cũng dẫn đến doanh thu cao.
* Ở các mức discount cao (0.4 đến 0.8), sales có xu hướng giảm hoặc thấp hơn. Điều này có thể do giảm giá mạnh gây ra suy giảm giá trị sản phẩm trong mắt khách hàng, hoặc khách hàng mục tiêu của sản phẩm có thể không quá nhạy cảm với mức discount.

Discount vs Profit :

* Profit (lợi nhuận) dường như giảm mạnh khi discount tăng. Các mức discount từ 0.4 trở lên hầu hết cho thấy profit âm, điều này có thể là dấu hiệu của việc chiết khấu quá mức dẫn đến lỗ.
* Ở mức discount thấp (0 đến 0.2), vẫn có nhiều điểm profit dương, cho thấy rằng mức discount nhẹ có thể giúp duy trì lợi nhuận. Điều này gợi ý rằng doanh nghiệp nên cân nhắc mức giảm giá vừa phải để vừa thu hút khách hàng vừa đảm bảo lợi nhuận.

=> Biểu đồ này cho thấy việc tăng mức giảm giá không hẳn sẽ giúp tăng doanh thu và ngược lại, có thể gây ảnh hưởng xấu đến lợi nhuận. Để đạt được lợi nhuận tốt, doanh nghiệp cần tìm được mức discount hợp lý thay vì áp dụng mức giảm giá cao.
